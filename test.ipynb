```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv(r'F:\practice\Python-Visualization\student_exam_scores.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
x = data['hours_studied']
y = data['exam_score']

plt.scatter(x, y)
plt.xlabel('Hours Studied')
plt.ylabel('Exam Score')
plt.title('Hours Studied vs Exam Score')
```

```python
x = data['sleep_hours']
y = data['exam_score']

plt.scatter(x, y)
plt.xlabel('Hours of Sleep')
plt.ylabel('Exam Score')
plt.title('Hours of Sleep vs Exam Score')
```

```python

x = data["attendance_percent"]
y1 = data["previous_scores"]
y2 = data["exam_score"]
plt.scatter(x, y1, label='Previous Scores', marker='o')
plt.scatter(x, y2, label='Exam Score', marker='o')
plt.xlabel('Hours Studied')
plt.ylabel('Scores')
plt.title('Previous Scores and Exam Score vs Hours Studied')
plt.legend()
```
