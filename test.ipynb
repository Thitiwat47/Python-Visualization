```python
import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv(r'F:\practice\Python-Visualization\house_prices.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
data.isnull().sum()
```

```python
d = data['Price (in rupees)']
d.fillna(d.median(), inplace=True)
```

```python

plt.boxplot(d)
plt.show()
```

```python
Q1 = d.quantile(0.25)
Q3 = d.quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

outliers = d[(d < lower_bound) | (d > upper_bound)]
print(outliers)
```

```python
df_no_outliers = d[(d >= lower_bound) & (d <= upper_bound)]
print(df_no_outliers)
```

```python


plt.boxplot(df_no_outliers)

plt.show()
```
